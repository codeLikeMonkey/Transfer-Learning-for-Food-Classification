```python
import torch
from torch import nn
import torch.nn.functional as F
from torch.autograd import Variable
import numpy as np
import glob
import os
import matplotlib.pyplot as plt
import torchvision.models as models
import cv2
```

```python
class Extractor(nn.Module):
    def __init__(self):
        super(Extractor, self).__init__()
        
        self.net1 = models.vgg16(pretrained=True)        
        self.net1.classifier = nn.Sequential(
            *(self.net1.classifier[i] for i in range(5)))
        
        self.net2 = models.vgg19(pretrained=True)        
        self.net2.classifier = nn.Sequential(
            *(self.net2.classifier[i] for i in range(5)))
        
        
    def forward(self, images): 
        
        out1 = self.net1(images)
        out2 = self.net2(images)
        return torch.cat((out1,out2),1)
```

```python
def read_img(img_path):
    img = plt.imread(img_path)
    img = cv2.resize(img,(224,224))
    img = img / 255
    img = img.transpose((2,0,1))
    return img
```

```python
with open("./food-101/meta/train.txt","r") as f:
    training_list = f.readlines()
with open("./food-101/meta/test.txt","r") as f:
    test_list = f.readlines()
    
with open('./food-101/meta/labels.txt','r') as f:
    labels = f.readlines()
    
training_list = ["./food-101/images/"+ path[:-1] +".jpg" for path in training_list]
test_list = ["./food-101/images/"+ path[:-1] +".jpg" for path in test_list]
labels = [label[:-1].replace(" ",'_').lower() for label in labels]
```

```python
net = Extractor()
net.cuda()
```

```python
import pickle

training_data = []
counter = 0
for i in range(len(training_list)):
    img = read_img(training_list[i])
    x = torch.FloatTensor(img).view(1,3,224,224)
    x = Variable(x.cuda())
    feature = net(x)
    label_id = labels.index(training_list[i].split("/")[-2])
    temp = (label_id,feature.cpu().data.numpy())
    training_data.append(temp)
    if i % 1000 == 0:
        print(i)
```

```python
del net
```

```python
 with open("training_data_vgg16vgg19_all.pkl","wb") as f:
        pickle.dump(training_data, f)
```
